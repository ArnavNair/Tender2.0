```python
military_x = military.iloc[:,0].values.reshape(-1,1)
military_y = military.iloc[:,1].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(military_x,military_y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(military_x,military_y,test_size=0.1)
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(50)
rf.fit(x_train,y_train)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_train,y_train)
plt.plot(x_train,rf.predict(x_train),'red')
plt.show
```

```python
y_pred = rf.predict(x_test)
```

```python
sc.fit_transform(capital)
sc.fit_transform(social_sector)
```

```python
capital_x = capital.iloc[:,0].values.reshape(-1,1)
capital_y = capital.iloc[:,1].values.reshape(-1,1)
social_sector_x = social_sector.iloc[:,0].values.reshape(-1,1)
social_sector_y = social_sector.iloc[:,1].values.reshape(-1,1)
```

```python
cx_train,cx_test,cy_train,cy_test = train_test_split(capital_x,capital_y,test_size=0.1)
scx_train,scx_test,scy_train,scy_test = train_test_split(social_sector_x,social_sector_y,test_size=0.1)
```

```python
rfc = RandomForestRegressor(50)
rfsc = RandomForestRegressor(50)
```

```python
rfc.fit(cx_train,cy_train)
rfsc.fit(scx_train,scy_train)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(cx_train,cy_train)
plt.plot(cx_train,rfc.predict(cx_train),'red')
plt.show
```

```python
plt.figure(figsize=(10,10))
plt.scatter(scx_train,scy_train)
plt.plot(scx_train,rfsc.predict(scx_train),'red')
plt.show
```

```python
cy_pred = rfc.predict(cx_test)
scy_pred = rfsc.predict(scx_test)
```
